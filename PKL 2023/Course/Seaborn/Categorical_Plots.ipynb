```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x ='sex', y='total_bill', data = tips, estimator=np.std )
#sns.barplot(x ='sex', y='total_bill', data = tips) defaultnya mean
```

```python
sns.countplot(x = 'sex', data = tips) #menghitung banyaknya jenis kelamin
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker') 
#bisa juga untuk membandingkan 3 category
#ingat kembali cara membaca boxplot
```

```python
#sns.violinplot(x='day', y='total_bill', data=tips) 
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split = True) 
```

```python
#sns.stripplot(x='day', y='total_bill', data=tips, jitter = False)
sns.stripplot(x='day', y='total_bill', data=tips, jitter = True, hue ='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black') #tidak ada yang ditimpa
```

```python
#sns.catplot(x='day', y='total_bill', data=tips, kind='bar') 
sns.catplot(x='day', y='total_bill', data=tips, kind='violin') 
#factorplot sudah renamed menjadi catplot
```
