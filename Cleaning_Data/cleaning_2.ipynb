```python
import pandas as pd
```

```python
import csv
```

```python
movie = pd.read_csv("/content/movies_metadata.csv")
```

```python
ratings = pd.read_csv('/content/ratings_small.csv')
```

```python
movie.columns
```

```python
movie = movie[['genres', 'id', 'original_title', 'tagline']]
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movie.dropna(inplace=True)
```

```python
# movie.count
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.rename(columns={'id': 'movieId'}, inplace=True)
```

```python
movie.head()
```

```python
ratings.count
```

```python
ratings.head()
```

```python
movie['movieId'] = movie['movieId'].astype(int)
```

```python
merged = movie.merge(ratings, on='movieId')
```

```python
merged.to_csv('movie.csv', index=True)
```
