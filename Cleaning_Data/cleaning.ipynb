```python
import pandas as pd
import numpy as np
```

```python
#reading dataset
movie=pd.read_csv('tmdb_5000_movies.csv')
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
credit.count()
```

```python
movie.count()
```

```python
movie=movie.merge(credit,on='title')
```

```python
movie.count()
```

```python
movie = movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie.count()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movie.dropna(inplace=True)
```

```python
movie.count()
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head(2)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head(2)
```

```python
def convert_actors(text):
    actors = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            actors.append(i['name'])
        counter+=1
    return actors

movie['cast'] = movie['cast'].apply(convert_actors)
movie.head(2)
```

```python
movie.count()
```

```python
def fetch_director(text):
    director = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            director.append(i['name'])
    return director

movie['crew'] = movie['crew'].apply(fetch_director)
movie.head(2)
```

```python
movie.count()
```

```python
movie.head()
```

```python
movie.count()
```

```python
output_file = 'movie_scikit.csv'
movie.to_csv(output_file, index=False, header=True)
```
